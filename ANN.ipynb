```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import History
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
df = pd.read_csv("data_set")
df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python

#shuffeling the data set
data = df_shuffled.values  

X = data[:, 1:6]  # entrée
y = data[:, 0]    # vérité de base 

# deviding the data set into training and test test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)


# normalisation of the data
scaler_X = MinMaxScaler()
X_train = scaler_X.fit_transform(X_train)
X_test = scaler_X.transform(X_test)

scaler_y = MinMaxScaler()
y_train = scaler_y.fit_transform(y_train.reshape(-1, 1)).flatten()
y_test = scaler_y.transform(y_test.reshape(-1, 1)).flatten()





# Model
model = Sequential()
model.add(Dense(40, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(40, activation='relu'))
model.add(Dense(1))

#using gradient descent as an optimization method
descente_de_gradient = tf.keras.optimizers.SGD(learning_rate=0.01)
model.compile(optimizer=descente_de_gradient, loss='mean_squared_error')

    
#training and printing the error of each epoch
history = model.fit(X_train, y_train, epochs=100, batch_size=32,validation_split=0.2, verbose=1)



```

```python
#plotting the error\cost with respect to epochs 
plt.plot(history.history['loss'], label='cout')
plt.plot(history.history['val_loss'])
plt.xlabel('Epochs')
plt.ylabel('cost / validation')
plt.title("error\cost with respect to epochs ")
plt.legend(['cost','validation'], loc='upper right')
plt.show()
```

```python
#testing the model

predictions = model.predict(X_test)
total_predictions = len(X_test)

# Reverse normalization for y_test and predictions
y_test_original = scaler_y.inverse_transform(y_test.reshape(-1, 1)).flatten()
predictions_original = scaler_y.inverse_transform(predictions).flatten()

# Define margin
margin = 0.5

# Count the number of correct predictions within the margin
correct_predictions = np.sum(np.abs(predictions_original - y_test_original) <= margin)

# Calculate accuracy
accuracy_within_margin = correct_predictions / total_predictions

# Print results
print(f"number of correct predictions : {correct_predictions}")
print(f"total number of predictions : {total_predictions}")
print(f"accuracy within the defined margin: {accuracy_within_margin * 100:.2f}%")
```

```python
#evaluating the model with RMSE, MAE, R2
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt


# Calculate metrics
rmse = sqrt(mean_squared_error(y_test, predictions))
mae = mean_absolute_error(y_test, predictions)


print(f'RMSE: {rmse}')
print(f'MAE: {mae}')
```
